```r
# Que1 Covariance of Time Reading and TimeTv

cov(df$TimeReading,df$TimeTV)
```

```r
#  The result indicates that TimeReading and TimeTV has an inverse relationship.
# As time spent on watching TV goes up, time spent on reading goes in the other direction.

# I will want to use this measure to investigate relationship before moving into invetigating causation.
```

```r
# Que ii.  It seems TimeReading is measured in hours while TimeTV is measured in minutes.  
# So the two variables are not measured using the same scale.  If this variables are standardized, 
# the result might change.
```

```r
# Que iii.  I will use correlation coefficient to perform this kind of analysis because the covariance results
# does not provide the magnitude of variation between the two variables.  With correlation analysis, one is able
# to guage the extent of relationship between the two variables.
```

```r
# Que iv. correlation

cor(df,use="complete.obs", method = "pearson")
```

```r
cor(df$TimeReading,df$Happiness,use="complete.obs",method="pearson")
```

```r
cor.test(df$TimeReading,df$Happiness,alternative="less",method="pearson",conf.level=0.99)
```
