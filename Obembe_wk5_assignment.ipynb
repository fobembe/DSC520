### Exploring filtering function

```r
# Filter the table for 'cut' = Premium

library("dplyr")
```

```r
df <- diamonds
head(df)
```

```r
prem<- filter(df, cut == "Premium")
head(prem)
```

```r
# Filter for color = 'J'

colr <- filter(df,color=="J")
head(colr)
```

### Using Select Function

```r
df1 = df %>% select(carat,cut,color, price)
head(df1)
```

```r
# Select another set of variables from the df table

df2 = df %>% select(depth,table,x,z)
head(df2)
```

### Using the Summarize function

```r
summarize(df,pc=mean(price))
```

```r
# Summarize the table column of df2

summarize(df2, TB=mean(table))
```

### Using the Groupby function

```r
df %>% 
group_by(cut) %>%
summarize(average_price = mean(price))
```

### Using arrange to order data

```r
# given our data frame df

head(df)
```

```r
# we wan to reorder the df in the order of price, carat, table

rangedf <- df %>% arrange(price,carat)
head(rangedf)
```

### Using mutate to add new variables

```r
# Add a total revenue column by multiplying price by depth

Total_Revenu = df %>% mutate(TR = price*depth)
head(Total_Revenu)
```

### Using RBIND and CBIND functions

```r
# Splitting the df dataset into 2 df_1 and df_2
df_1 <- df %>% select(carat,cut,color,clarity,depth,table)
df_2 <- df %>% select(price,x,y,z)
```

```r
head(df_1)
```

```r
head(df_2)
```

```r
# cbind df_2 to df_1

new_df <- cbind(df_1,df_2)
head(new_df)
```

```r
# filter some rows and rbind it to new_df

prem_df <- filter(df,cut=="Premium")
head(prem_df)
```

```r
new_df2 <- rbind(new_df,prem_df)
```

```r
str(new_df)
str(new_df2)
```
